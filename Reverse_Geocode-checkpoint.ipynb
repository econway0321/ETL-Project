```python
#Import Dependencies
import pandas as pd
import reverse_geocoder as rg 
import pprint 
```

```python
#Read into cleaned csv file with lat/lon coordinates
coor_df = pd.read_csv('facility_latlon_data.csv')
coor_df
```

```python
#Define search for per coordinate info
def reverseGeocode(coordinates): 
    coord_info = rg.search(coordinates)
    return (coord_info)

if __name__=="__main__": 

    coordinates =list(zip(coor_df['FacilityLatitude'],coor_df['FacilityLongitude'])) 
    data = reverseGeocode(coordinates)

    coor_df['name'] = [city['name'] for city in data]
    coor_df['admin1'] = [city['admin1'] for city in data]

coor_df.to_csv("data_appended.csv") 
```

```python
#Read into appended csv file
check = pd.read_csv("data_appended.csv")
check.head()
```

```python
#Rename columns to maintain format
rename = check.rename(columns={'name': 'Name', 'admin2': 'County'})
```

```python
#Drop columns that aren't needed
new_df = rename.drop(['Unnamed: 0', 'admin1'], axis=1)
```
