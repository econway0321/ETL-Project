```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
#2010
```

```python
df_2010 = pd.read_csv('2010.csv', encoding= 'cp1252')
```

```python
list_columns = list(df_2010.columns)
list_columns
```

```python
select_columns = ['OrderNumber', 'Park', 'SiteType', 'FacilityID', 'FacilityState', 'StartDate', 'EndDate', 'NumberOfPeople']
new_df10 = df_2010.loc[:, select_columns]
```

```python
filter_10CO = new_df10[new_df10['FacilityState'] == 'CO']
filter_10CO.head()
```

```python
list(filter_10CO['SiteType'])
```

```python
site10list = ['CABIN ELECTRIC', 'CABIN NONELECTRIC', 'EQUESTRIAN NONELECTRIC', 'GROUP PICNIC AREA', 'GROUP SHELTER ELECTRIC',
            'GROUP SHELTER NONELECTRIC', 'GROUP STANDARD AREA NONELECTRIC', 'GROUP STANDARD ELECTRIC', 'GROUP STANDARD NONELECTRIC',
            'GROUP TENT ONLY AREA NONELECTRIC', 'GROUP WALK TO', 'LOOKOUT', 'RV ELECTRIC', 'STANDARD ELECTRIC', 'STANDARD NONELECTRIC', 
            'TENT ONLY NONELECTRIC', 'WALK TO']
return10 = filter_10CO[filter_10CO['SiteType'].isin(site10list)]
return10.head()
```

```python
check = return10[return10['SiteType'] == 'Combo Tour']
check
```

```python
return10.to_csv('clean_res10', index=False, header=True)
```
