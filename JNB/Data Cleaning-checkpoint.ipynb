```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
#Read into CSV for 2016 data
df_2016 = pd.read_csv('2016.csv', encoding='cp1252')
```

```python
#Get a list of column headers to reorganize the data
list_columns = list(df_2016.columns)
# list_columns
```

```python
#Select desired columns and reorganize data
select_columns = ['OrderNumber', 'Park', 'SiteType', 'FacilityID', 'FacilityState', 'StartDate', 'EndDate', 'NumberOfPeople']

new_df16 = df_2016.loc[:, select_columns]
# new_df16
```

```python
#General check of information
# new_df16.count()
# len(new_df16)
```

```python
#Filter data for Colorado only
filter_16CO  = new_df16[new_df16['FacilityState'] == 'CO']
# filter_16CO.head()
```

```python
#Suspect that 'SiteType' data is including more than just overnight bookings, i.e. 'Historic Tour' = 'activity'
#Therefore, list value types in column 'SiteType' to filter our activity like bookings
list(filter_16CO['SiteType'])
```

```python
#Construct a list which specifies the type of data values desired for SiteType
site16List = ['CABIN ELECTRIC', 'CABIN NONELECTRIC', 'EQUESTRIAN NONELECTRIC', 'GROUP PICNIC AREA', 'GROUP SHELTER ELECTRIC',
            'GROUP SHELTER NONELECTRIC', 'GROUP STANDARD AREA NONELECTRIC', 'GROUP STANDARD ELECTRIC', 'GROUP STANDARD NONELECTRIC',
            'GROUP TENT ONLY AREA NONELECTRIC', 'GROUP WALK TO', 'LOOKOUT', 'RV ELECTRIC', 'STANDARD ELECTRIC', 'STANDARD NONELECTRIC', 
            'TENT ONLY NONELECTRIC', 'WALK TO']

#Use the .isin() method to filter/set specified values for SiteType column
return16 = filter_16CO[filter_16CO['SiteType'].isin(site16List)]
# return16
```

```python
#Check to see if SiteType returns unwanted values
# check = return16[return16['SiteType'] == 'Combo Tour']
# check
```

```python
#If check passes, save into a clean CSV
return16.to_csv('clean_res16', index=False, header=True)
```
